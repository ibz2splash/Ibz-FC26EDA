## Key Findings

- **16,228 players** across 59 attributes analyzed
- **England** has the most represented players (1,400+), followed by Germany and Argentina
- **Mohamed Salah** is the highest rated player in EAFC 26 (OVR 91)
- **Arsenal** has the highest average squad rating among all clubs
- Most players peak between **ages 24–26**, with sharp decline after 33
- **Passing and Dribbling** are the most correlated stats (0.86)
- **Defenders genuinely can't shoot** — DEF vs SHO correlation is -0.40
- **23 hidden gems** identified: age ≤ 21 with OVR ≥ 78, led by Lamine Yamal (89)
- **Lamine Yamal (18)** and **Estêvão (18)** are the standout young talents
